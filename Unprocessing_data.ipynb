```python
import numpy as np
from PIL import Image
import os

# Path to the .npy file
file_path = '/Users/rohith/Desktop/M3D copy/Data/data/examples/example_06.npy'

# Load the .npy file
data = np.load(file_path)

# Remove the first dimension (batch dimension), now shape is (32, 256, 256)
slices = data[0]  # Shape: (32, 256, 256)

# Directory to save the output images
output_dir = 'output_images'
os.makedirs(output_dir, exist_ok=True)

# Loop through each slice and save it as a PNG
for i in range(slices.shape[0]):
    # Normalize the slice data (convert float32 to 0-255 range)
    slice_data = slices[i]
    slice_normalized = (slice_data - np.min(slice_data)) / (np.max(slice_data) - np.min(slice_data)) * 255
    slice_normalized = slice_normalized.astype(np.uint8)  # Convert to uint8

    # Convert to an image and save as PNG
    img = Image.fromarray(slice_normalized)
    output_path = os.path.join(output_dir, f'slice_{i}.png')
    img.save(output_path)
    print(f'Slice {i} saved as {output_path}')
```
